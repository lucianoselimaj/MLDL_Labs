```python
import matplotlib.pyplot as plt

# Step 1: Create a figure with 2 rows and 3 columns of subplots
fig, axes = plt.subplots(2, 3, figsize=(12, 6))  # Optional: figsize controls overall size

# Step 2: Loop through each Axes to set labels and titles
for i in range(2):          # rows
    for j in range(3):      # columns
        ax = axes[i, j]     # Get the subplot at position (i, j) --> selects each subplot

        # Example: plot something (optional)
        ax.plot([0, 1], [i+j, i+j+1])

        # Set x-axis label and title
        ax.set_xlabel(f"x-axis ({i},{j})")
        ax.set_title(f"Title ({i},{j})")

# Step 3: Adjust layout to avoid overlapping labels
plt.tight_layout()

# Step 4: Show the whole figure
plt.show()
```

# **2. Object oriented**:
instead of using automatic plotting like plt.plot(), this method gives us more control. We can create the plot objects(figure and axes), and then use their methods to draw and customize. <br><br>

this is especially useful when we're:
*   Creating multiple charts in one figure
*   Adding annotations or shapes
*   Making comple or high customized plots



```python
import matplotlib.pyplot as plt



#Firs of all create the figure and axes(subplots)
'''
1. fig is the Figure object (the canvas)
2. ax is the Axes object (the chart)

Now you can control them separately:
fig = the full window
ax = one specific chart area inside that window
'''
fig, ax = plt.subplots()

# Plot data on the axis
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 16]


#Optional: we can set the figure size
fig.set_size_inches(10, 10)

#draw the plot, remember that we are using the obj oriented method so we plot using the ax not the figure
ax.plot(x,y)

#set title and labels of the provided axes
ax.set_title("Simple Line Plot")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")

#show the plot
plt.show()
```

# **SUBPLOTS**
we use subplots when we want to show more than one chart in the same figure(canvas). <br>
We can think it as a grid within the figure (like rows and columns), where each cell in the grid is a separate chart (axes)

```python
import matplotlib.pyplot as plt

#we crate a 2-row and 2-column grid of subplots, so in total we have 4 charts
# fig is the cnavas
#ax is like a 2d array (ax[0,0],ax[0,1], etc...)
fig, ax = plt.subplots(2,2, figsize=(6,4))

#draw subplot charts
ax[0,0].plot([0, 1, 2], [2, 4, 6])
ax[1,1].plot([0, 1, 2], [3,6,9])

#automatically adjusts the layout so charts don't overlap each other
plt.tight_layout()

plt.show()
```

# **Custom Subplots layout**

# **1. plt.subplot_mosaic()**
while plt.subplots() makes a grid like 2x2, sometimes we want a layout that isn't **perfectly rectangular**. <br><br>
For example, maybe:
*   One plot should be big across the top
*   Two smaller plots below it

that's where .subplots_mosaic() helps

```python
import matplotlib.pyplot as plt

#this creates a custom layout
fig, ax = plt.subplot_mosaic("""
AAABB
AAABB
AAACD
""", figsize=(5, 3))

'''
Layout Diagram (rows of characters):
Row 1: A A A B B
Row 2: A A A B B
Row 3: A A A C D

"A" fills the big top-left area (3 wide, 3 tall)
"B" is to the right of A (2 wide, 2 tall)
"C" and "D" are small plots at the bottom right
'''

#plot a line chart in the section A
ax["A"].plot([1, 2, 3], [1, 2, 3])
#plots a line chart in the section B
ax["B"].plot([1, 2, 3], [3, 2, 1])
ax["C"].plot([1, 2, 3], [-1, 1, -1])
ax["D"].plot([1, 2, 3], [1, -1, 1])

#auto-adjusting
plt.tight_layout()
plt.show()
```

# **Legend**
useful when we draw multiple lines, so you know which is which

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 15, 25, 30]
y2 = [30, 25, 20, 15, 10]


fig, ax = plt.subplots()

#First line
ax.plot(x, y1, label='Line A', color='blue')
#second line
ax.plot(x, y2, label='Line B', color='green')

ax.set_title('Graph. with Legend')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')

# Shows the legend box
ax.legend()
#add the background grid
ax.grid(True)

plt.show()
```

# **Customizing Colors, Styles, and Markers**

# **Subplots â€” Showing Multiple Charts in One Figure**
Earlier we saw how subplot_mosaic() allows flexible layouts. Now letâ€™s look at how to create multiple subplots using plt.subplots(), the standard grid approach.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

# Create a figure with 1 row and 2 columns of subplots
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(14,7))

# First subplot: Line chart
ax1.plot(x,y, color='blue', linestyle='--', marker='*')
ax1.set_title('Line plot')

ax2.scatter(x,y, color='red')
ax2.set_title('Scatter plot')

plt.tight_layout()
plt.show()
```

# **Annotations and Text in Matplotlib**
annotations let you write text directly on the chart, next to the specific points, to explain what they mean.

Mainly used for:
* Highlight an important point
* Show a value
* Add a label to specific location

# **Basic function: ax.text(x,y, 'text', color='...')**
we choose:
* where to place the text --> (x,y)
* what the text says --> 'some label'
* What color to make it --> color='red'

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

#build  the subplotting
fig, ax = plt.subplots()

#draw the line
ax.plot(x,y)

#Annotate the specific point (x=3,y=15)
ax.text(3,15, 'Important Point', color='red')

#set the fields --> remember that we are working with ax
ax.set_title("Chart with Annotation")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")

plt.show()
```

We can also use ax.annotate() to draw arrows pointing to data

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

fig, ax = plt.subplots()

ax.plot(x,y)

ax.annotate("This peak", xy=(4, 25), xytext=(3, 28),
            arrowprops=dict(facecolor='black', arrowstyle='->'))

#set the fields --> remember that we are working with ax
ax.set_title("Chart with Annotation")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")

plt.show()
```

# **Colormaps**
a colormap gradient of colors used to represent data **intensity**, **heat**, or **categories**. <br><br>
They are very useful in:
* heatmaps
* Scatter Plots(color by value)
* 3D surface plots
* Image data

# **Heatmap**

```python
import numpy as np
import matplotlib.pyplot as plt

# Create random 2D data (10x10 grid) of random numbers between 0 and 1
data = np.random.rand(10,10)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(6,5))

# Show the data as a heatmap using the "viridis" colormap
im = ax.imshow(data, cmap='viridis')

# Add a color bar (legend for colors)
cbar = ax.figure.colorbar(im, ax=ax) #adds the colorbar to the figure, using the heatmap im, and placing the colorbar relative the the main axes
cbar.ax.set_ylabel('Value', rotation=-90, va="bottom")#sets the label the the colorbar, and then rotates the text vertically. Finally alligns the text using bottom

plt.title("Heatmap with Colormap")
plt.show()
```

A heatmap helps you:
* Quickly spot patterns or anomalies
* Compare which areas have higher or lower values
* See clusters, gradients, and trends

These show the highest and lowest values. <br>
Find any square thatâ€™s bright **yellow** â†’ those are your highest values (close to 1).<br>
Find any square thatâ€™s dark **purple** â†’ those are your lowest values (close to 0).
